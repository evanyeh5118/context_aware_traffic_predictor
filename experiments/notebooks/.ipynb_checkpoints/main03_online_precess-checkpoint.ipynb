```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from dataclasses import dataclass

import json
import sys
import numpy.random as random
sys.path.append("../../")

import time 

from src.online_prediction.OnlineDataProcessor.dataProcessor import DataProcessor
```

```python
from pathlib import Path

# Use absolute path for reliability in Jupyter notebooks
project_root = Path(r"c:\Users\Ye\Documents\YuYeh_Documents\L2S\Projects\context_aware_traffic_predictor")
csv_path = project_root / "data" / "processed" / "dpdr" / "combined_forward.csv"

csv_data = pd.read_csv(csv_path)
print(csv_data.head(5))
```

```python
@dataclass
class config:
    dim_data = 12
    window_length: int = 20
    history_length: int = 100
    smooth_fc: float = 1.5
    smooth_order: int = 3
    Ts : float = 0.01
    min_vals: float = -5.0*np.ones(dim_data)
    max_vals: float = 5.0*np.ones(dim_data)


dataProcesoor = DataProcessor(config)

#sine_data = np.column_stack((np.sin(np.arange(0, 10, 0.1)), np.cos(np.arange(0, 10, 0.1))))
#print(sine_data.shape)

last_time = time.time()

context_list = []
context_no_smooth_list = []
traffic_list = []
count = 1
N_count = 100
while True:
    if time.time() - last_time > config.Ts:
        #data_point = sine_data[count, :]
        transmission = csv_data.iloc[count, 1]
        data_point = csv_data.iloc[count, 2:].values
      
        if int(transmission) == 1:
            dataProcesoor.add_data_point(data_point)
        last_time = time.time()

        if count % config.window_length == 0:
            #context_no_smooth, flags, timestamps_bin = dataProcesoor.get_window_data()
            context, last_trans_sources, context_no_smooth, debugs = dataProcesoor.get_window_features()
            (flags, timnestamps_bin) = debugs
            context_list.append(context)
            context_no_smooth_list.append(context_no_smooth)
            traffic_list.append(np.sum(flags))
        count += 1
    
    if count >= N_count:
        break

dim_data = 12
context_list = np.array(context_list).reshape(-1, dim_data)
context_no_smooth_list = np.array(context_no_smooth_list).reshape(-1, dim_data)
traffic_list = np.array(traffic_list).reshape(-1)

print(context_list.shape)
print(context_no_smooth_list.shape)
plt.plot(context_list[: , 0])
plt.plot(context_no_smooth_list[:, 0])
#plt.plot(csv_data.iloc[:N_count, 2].values)
plt.show()
plt.plot(traffic_list)
plt.show()
```

```python
from src.context_aware.preprocessing.Helpers import smoothDataByFiltfilt

history_context = np.vstack(np.array(dataProcesoor.history_context))
print(history_context.shape)
history_context_smooth = smoothDataByFiltfilt(
    history_context, dataProcesoor.smooth_fc, dataProcesoor.smooth_fs, dataProcesoor.smooth_order)

plt.plot(history_context[:,0])
plt.plot(history_context_smooth[:,0])
plt.show()
```
